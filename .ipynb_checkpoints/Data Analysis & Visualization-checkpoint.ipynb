```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv(r'dataset.csv')
```

```python
dataset.info()
```

```python
dataset.describe().T
```

```python
# here BMI Stands for Body Mass Index indicating the level of obesity (ranging from 15 to 40).
```

```python
sns.countplot(x='Sex', data=dataset )
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Distribution')
plt.show()
print('Male Count : ',dataset['Sex'].value_counts()[1])
print('Female Count : ',dataset['Sex'].value_counts()[0])
print(dataset['Sex'].value_counts())
```

```python
sns.countplot(x='Smoker', data=dataset )
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.title('Smoker Distribution')
plt.show()
print('Yes Count : ',dataset['Smoker'].value_counts()[1])
print('No Count : ',dataset['Smoker'].value_counts()[0])
print(dataset['Smoker'].value_counts())
```

```python
sns.countplot(x='Region', data=dataset , color='r')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region Distribution')
plt.show()
print('northwest Count : ',dataset['Region'].value_counts()[0])
print('southwest Count : ',dataset['Region'].value_counts()[1])
print('southeast Count : ',dataset['Region'].value_counts()[2])
print('northeast Count : ',dataset['Region'].value_counts()[3])
print(dataset['Region'].value_counts())
```

```python
sns.heatmap(dataset.corr() , cmap= 'Blues' , annot=True , annot_kws={'size':8} )
plt.title('Heatmap of Correlation Factors')
plt.show()
```
